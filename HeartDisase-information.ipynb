```python
import matplotlib.pyplot as plt
# Group by 'sex' and 'target' and count occurrences
count_data = df.groupby(['sex', 'target']).size().unstack()

# Plot a bar chart
count_data.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Sex (1: Female, 0: Male)')
plt.ylabel('Count')
plt.title('Distribution of Target by Sex')

# Show the plot
plt.show()
```

```python
sns.histplot(data=df, x='age', kde=True)
```

```python
sick_women = df[(df['sex'] == 1) & (df['target'] == 1)].shape[0]
sick_men = df[(df['sex'] == 0) & (df['target'] == 1)].shape[0]

# Create a pie chart
labels = ['Patient Women', 'Patient Men']
sizes = [sick_women, sick_men]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Make the pie chart circular
plt.title('Percentage of Sick Women and Men')
plt.show()
```

```python
sns.histplot(df['age'][df['target']==1], kde=True, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution of Sick Patients')

# Show the plot
plt.show()
```
