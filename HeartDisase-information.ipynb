```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Group by 'sex' and 'target' and count occurrences
count_data = df.groupby(['sex', 'target']).size().unstack()

# Plot a bar chart
count_data.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Sex (1: Female, 0: Male)')
plt.ylabel('Count')
plt.title('Distribution of Target by Sex')

# Show the plot
plt.show()
```
