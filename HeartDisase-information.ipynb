age: Hastanın yaşı.

sex: Hastanın cinsiyeti (1 = erkek, 0 = kadın).

cp: Göğüs ağrısı tipi (0 = tipik anjina, 1 = atipik anjina, 2 = anjinal olmayan ağrı, 3 = asemptomatik).

trestbps: Hastanın hastaneye başvurduğunda dinlenirken kan basıncı (mm Hg cinsinden).

chol: Serum kolesterol düzeyi (mg/dl cinsinden).

fbs: Açlık kan şeker düzeyi > 120 mg/dl (1 = doğru, 0 = yanlış).

restecg: Dinlenme elektrokardiyografik sonuçlar (0 = normal, 1 = ST-T dalga anormalliği, 2 = muhtemel veya kesin sol ventriküler hipertrofi).

thalach: Ulaşılan maksimum kalp atış hızı.

exang: Egzersize bağlı anjina (1 = evet, 0 = hayır).

oldpeak: Dinlenme durumuna göre indüklenen egzersiz sonrası ST depresyonu.

slope: Zirve egzersiz ST segmentinin eğimi (0 = aşağı eğimli, 1 = düz, 2 = yukarı eğimli).

ca: Floroskopi ile renklendirilmiş büyük damar sayısı (0-3).

thal: Talasemi - bir kan bozukluğu (3 = normal; 6 = sabit kusur; 7 = düzeltilen kusur).

target: Kalp hastalığı varlığı (1 = kalp hastalığı var, 0 = kalp hastalığı yok).

Average-Age: Muhtemelen türetilmiş veya hesaplanmış bir sütun, veri setindeki yaşın ortalamasını temsil ediyor.

Average_Of_Women: Muhtemelen başka bir türetilmiş veya hesaplanmış bir sütun, veri setindeki kadınların bir değişkenin ortalamasını temsil ediyor.

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Group by 'sex' and 'target' and count occurrences
count_data = df.groupby(['sex', 'target']).size().unstack()

# Plot a bar chart
count_data.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Sex (1: Female, 0: Male)')
plt.ylabel('Count')
plt.title('Distribution of Target by Sex')

# Show the plot
plt.show()
```
